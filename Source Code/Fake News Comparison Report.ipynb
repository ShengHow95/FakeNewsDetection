# Comparison between TFIDF and Sequence Method

```python
import json
import numpy as np
import matplotlib.pyplot as plt
```

### Data Import

```python
with open('Model_TFIDF_Title.json', 'r') as Model_TFIDF_File:
    tfidf_model_title = json.load(Model_TFIDF_File)
    
with open('Model_TFIDF_Content.json', 'r') as Model_TFIDF_File:
    tfidf_model_content = json.load(Model_TFIDF_File)
    
with open('Model_Sequence_Title.json', 'r') as Model_Sequence_File:
    sequence_model_title = json.load(Model_Sequence_File)

with open('Model_Sequence_Content.json', 'r') as Model_Sequence_File:
    sequence_model_content = json.load(Model_Sequence_File)
```

```python
Time = []
Accuracy = ['Title_Accuracy','Content_Accuracy']
Recall = ['Title_Recall','Content_Recall']
```

### Accuracy Comparison

```python
# create plot
fig, ax = plt.subplots(figsize=[5,5])

x_values = ['Model 1', 'Model 2', 'Model 3', 'Model 4']
y_values = [
            float(tfidf_model_title.get('Title_Accuracy'))*100, 
            float(tfidf_model_content.get('Content_Accuracy'))*100, 
            float(sequence_model_title.get('Title_Accuracy'))*100, 
            float(sequence_model_content.get('Content_Accuracy'))*100
           ]

plt.bar(x_values, y_values, color=['r','g','b','y'])
plt.xlabel('Keras Neural Network Model')
plt.ylabel('Performance (%)')
plt.ylim(70,100)
plt.title('Accuracy Comparison')

for i in range(4):
    ax.text(i - 0.25, y_values[i] + 1, "%.1f%%" % y_values[i], fontsize=12)

plt.show()
```

### Recall Comparison

```python
# create plot
fig, ax = plt.subplots(figsize=[5,5])

x_values = ['Model 1', 'Model 2', 'Model 3', 'Model 4']
y_values = [
            float(tfidf_model_title.get('Title_Recall'))*100, 
            float(tfidf_model_content.get('Content_Recall'))*100, 
            float(sequence_model_title.get('Title_Recall'))*100, 
            float(sequence_model_content.get('Content_Recall'))*100
           ]

plt.bar(x_values, y_values, color=['r','g','b','y'])
plt.xlabel('Keras Neural Network Model')
plt.ylabel('Performance (%)')
plt.ylim(80,100)
plt.title('Recall Comparison')

for i in range(4):
    ax.text(i - 0.25, y_values[i] + 0.5, "%.1f%%" % y_values[i], fontsize=12)

plt.show()
```
